# SinPlot()

Plotting the sinplot() with styles "whitegrid", "white", "ticks".

```python
%matplotlib inline

import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns

np.random.seed(sum(map(ord, "aesthetics")))

def sinplot(flip=1):
    x = np.linspace(0, 14, 100)
    for i in range(1, 7):
        plt.plot(x, np.sin(x + i * .5) * (7 - i) * flip)    
```

```python
sns.set_style("whitegrid")
sinplot()   
```

```python
sns.set_style("dark")
sinplot()
```

```python
sns.set_style("ticks")
sinplot()
```

```python
sinplot()
sns.despine()
```

```python
with sns.axes_style("darkgrid"):
    plt.subplot(211)
    sinplot()
plt.subplot(212)
sinplot(-1)
```
